# Data Analysis with Pandas

```python
# importing the pandas library 
import pandas as pd
```

```python
# Read a csv file from a local directory and load it into a dataframe (df)
df = pd.read_csv("./datasets/Gapminder.csv", sep=";")
df
```

```python
# Show first five rows of the dataframe
df.head()
```

```python
# Rename colums of the dataframe
colum_names = {
    "country": "COUNTRY",
    "continent": "CONTINENT",
    "year": "YEAR",
    "lifeExp": "LIFE_EXP",
    "pop": "POPULATION",
    "gdpPercap": "GDP_PERCAP"
}
df = df.rename(columns=colum_names)
```

```python
# Show fist 10 lines
df.head(10)
```

```python
# Get total rows and columns
df.shape
```

```python
# Get all column names
df.columns
```

```python
# Get type of data of each column
df.dtypes
```

```python
# Get the as five rows of the dataframe
df.tail()
```

```python
# Get descriptive information about the dataframe
df.describe()
```

```python
# Get unique values of a column
df["CONTINENT"].unique()
```

```python
# Filter the dataframe from a column value
df_oceania = df.loc[df["CONTINENT"] == "Oceania"]
print(df_oceania["CONTINENT"].unique())
df_oceania.head()
```

```python
# Group a dataframe by column values and get the number of values of other column
df.groupby("CONTINENT")["COUNTRY"].nunique()
```

```python
# What is the average life expectative by year
df.groupby("YEAR")["LIFE_EXP"].mean()
```

```python
# Get the average of the values of a column
df.GDP_PERCAP.mean()
```
